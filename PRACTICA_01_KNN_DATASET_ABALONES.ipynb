K-Nearest Neightbors

```python
#Primer Paso
# Definir las distancias de los Vectores de las Variables Independientes
X= abalones.drop("Rings", axis=1)
x= X.values # Variables Independientes

Y= abalones["Rings"]
y=Y.values # Variables Dependientes

# Aplicando un valor para K=3
K= 3

# Datos de un nuevo especimen de Abalon: Length= 0.569552, Diameter= 0.446407, Heigth= 0.154437, Whole Weight= 1.016849, Shucked Weight= 0.439051, Viscera Weight= 0.222526 &  Shell Weight= 0.291208

nuevos_datos={"Length": 0.569552, "Diameter": 0.446407, "Heigth": 0.154437, "Whole_Weight": 1.016849, "Shucked_Weight": 0.439051, "Viscera_Weight": 0.222526 , "Shell_Weight": 0.291208}

datos= [nuevos_datos["Length"], nuevos_datos["Diameter"], nuevos_datos["Heigth"], nuevos_datos["Whole_Weight"], nuevos_datos["Shucked_Weight"], nuevos_datos["Viscera_Weight"], nuevos_datos["Shell_Weight"]]

nuevo_punto_especimen= np.array(datos)

# Segundo Paso
# Calcular las distancias entre el nuevo punto y cada uno de los puntos que contiene el DataSetde Abalones, con el siguiente código:

distancias= np.linalg.norm(x-nuevo_punto_especimen, axis=1)

# Ahora con las  distancias entre vectores, necesitamos buscar a los vecinos más cercanos (K-NN) 
#Podemos usar el método llamado .argsort() para ordenar la matriz de menor a mayor, y puedes tomar los primeros k elementos para obtener los índices de los k vecinos más cercanos:

nearest_neighbor_ids= distancias.argsort()[:K] # Output: array([4045, 1902, 1644], dtype=int64), es la posición de los puntos en el DataFrame va a servir en un futuro para la prediccion de de la edad

print (f"Primeros tres vecinos cecanos:{nearest_neighbor_ids}")

# Promedio de Vario Vecinos
# Teniendo Identificados los indices de los tres vecinos mas cercanos del abalon que desconocemos la edad, ahora necesita combinar esos vecinos en una predicción para su nuevo punto de datos.
# Buscaremos los verdaderos vecinos cervanos:

nearest_neighbor_ids=Y[nearest_neighbor_ids]

print(f"Verdaderos Vecinos Cercanos:\n {nearest_neighbor_ids}")

# Combinaremos los multiples en vecinos en una predicción:

predicction= nearest_neighbor_ids.mean()

print (f"Prediccion:{predicction}")

# Vamosa partir la informacion en información en entrenamiento y en información de testeo.
# Usamos el método scikit learn:

X_train,X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=12345)
```

```python
# Método KNeighborsClassifier

KNN= KNeighborsClassifier(n_neighbors=1)

X,Y= abalones.loc[:,abalones.columns != "Rings"], abalones.loc[:,"Rings"]

KNN.fit(X_train, Y_train)

predicction= KNN.predict(X_test)


print ("predicction:{}".format(predicction))
print ("KNN classifier Whit KNN(K=3) accuracy is:", KNN.score(X_test,Y_test))

# Método adecuación de  KNN Regresion con  Scikit-learn:

# Este modelo usa NN (Nearest Neighbors) para predecir los valores de algún punto futuro

KNN_model= KNeighborsRegressor(n_neighbors=3)

KNN_model.fit(X_train, Y_train)

#Knn_model contiene todo lo necesario para hacer predicciones sobre nuevos puntos de datos de abulón.

prediction= KNN_model.predict(X_test)

print (f"KNN Regression Whit KNN(K=3) accuracy is:{KNN_model.score(X_test, Y_test)}")

```

```python
# Error Cuadrático Medio (RMSE:root-mean-square error).

train_preds= KNN_model.predict(X_train)

mse= mean_squared_error (Y_train, train_preds) # Desviación Cuadrática Media (MSE: mean squared deviation)

rmse= sqrt(mse)

print (f"Error Medio Cuadrado empleando KNN train:{rmse}")

# Tambien se puede usar RMSE con KNN Regression

test_preds= KNN_model.predict(X_test)

mse_KNN_REG= mean_squared_error (Y_test,  test_preds)

rmse_KNN_REG= sqrt(mse_KNN_REG)

print (f"Error Cuadrático Medio empleando KNN regression test:{rmse_KNN_REG}")

# Mediante el Error Cuadrático Medio se han generado dos medidas de vida de los abulones. En el primero se obtiene 1.65 años y en el Segundo se obtiene 2.37 años.

# Aún hace falta realizar ajustes de los hiperparametros para evitar los sobreajustes de los datos de entrenamiento.

```

Afinado y Optimizacion K-Nearest Neighbors usando Scikit Learn.

```python
# segundo Paso
# Mejorar el rendimiento de kNN en scikit-learn utilizando GridSearchCV.

# Vamos a ir probando diferentes valores de K (n_neighbors)=> en rangos de 1 a 50 para esta practica en el dataset de los abalones.

parameters= {"n_neighbors": range(1,50)}

gridsearch= GridSearchCV(KNeighborsRegressor(),parameters)

gridsearch.fit(X_train, Y_train)

gridsearch.best_params_
```

```python
# Ahora con el mejor valor de K=3 para el método KNN podremos observar la mejora de este mismo.

train_preds_grid= gridsearch.predict(X_train)

train_mse= mean_squared_error(Y_train, train_preds_grid)

train_rmse= sqrt(train_mse)

test_preds_grid= gridsearch.predict(X_test)

test_mse= mean_squared_error(Y_test, test_preds_grid)

grid_test_rmse= sqrt(test_mse)

print ("Error Cuadrático Medio empleando KNN regression empleando gridsearch train=",train_rmse)

print ("Error Cuadrático Medio empleando KNN regression empleando gridsearch test:",grid_test_rmse)
```

```python
#Precision del Método
gridsearch.score(X_test, Y_test)

print (f"KNN Regression Whit KNN Gridsearch(K=3) accuracy is:{gridsearch.score(X_test, Y_test)}")
```

```python
# Mejorar el rendimiento de kNN en scikit-learn utilizando GridSearchCV empleando la función weight function used in predition.

parameters= {"n_neighbors": range(1,50),"weights":["uniform",  "distance"]}

#Empleamos KNeighborsRegressor que fue donde obtuvimos un mejor valor de Error Cuadrático Medio empleando KNN regression:2.375417924000521

gridsearch_afinado= GridSearchCV(KNeighborsRegressor(),parameters)

gridsearch_afinado.fit(X_train, Y_train)

# Al final vamos a retener el mejor valor de K que se  pueda emplear 

gridsearch_afinado.best_params_

# Podemos observar que el mejor valor es K= 25

print (f"El mejor valor de K=> {gridsearch_afinado.best_params_}")

# Optimizando y Afinando K empleando el peso de los abalones

test_preds_grid_afinado=gridsearch_afinado.predict(X_test)

test_mse_afinado= mean_squared_error(Y_test, test_preds_grid_afinado)

grid_weight_test_rmse= sqrt(test_mse_afinado)

print(f"El valor Afinado y Optmizado del Error Cuadrático Medio empleando KNN regression (Gridsearch) con peso de los abulones: {grid_weight_test_rmse}")

# Aplicando gridsearch y buscando el mejor valor de K para el DataSet podemos observar una mejora en el error de predicción de la edad de los abulones de 2.17 a 2.1634

#Precision del Método
gridsearch_afinado.score(X_test, Y_test)

print (f"KNN Regression Whit KNN Gridsearch(K=3) accuracy is:{gridsearch_afinado.score(X_test, Y_test)}")

```

```python
# Tercer paso del afinado y optimización del método KNN es usando el embolsamiento (bagging).

#A) Create the KNeighborsRegressor with the best choices for k and weights that you got from GridSearchCV:

best_k= gridsearch_afinado.best_params_["n_neighbors"]

best_weights= gridsearch_afinado.best_params_["weights"]

bagged_knn= KNeighborsRegressor(n_neighbors=best_k, weights=best_weights)

#B) Import the BaggingRegressor class from scikit-learn and create a new instance with 100 estimators using the bagged_knn model:

bagging_model= BaggingRegressor(bagged_knn, n_estimators=100)

#Hay que emplear .fit() en los metodos
bagging_model.fit(X_train,Y_train)

#C) test_preds_grid_afinado= bagging_model.predict(X_test)

test_mse_bagging= mean_squared_error(Y_test,test_preds_grid_afinado)

bagging_test_rmse= sqrt(test_mse_bagging)

print(f"El valor Afinado y Optmizado del Error Cuadrático Medio empleando KNN regression (Bagging): {bagging_test_rmse}")

#El error de predicción en el kNN embolsado es 2,1616, que es ligeramente menor que el error anterior que obtuvo. Se necesita un poco más de tiempo para ejecutarlo, pero para este ejemplo, eso no es problemático.

#Precision del Método
bagging_model.score(X_test, Y_test)

print (f"KNN Regression Whit KNN Gridsearch(K=3) accuracy is:{bagging_model.score(X_test, Y_test)}")
```

La complejidad del Modelo, se puede mejorar el rango de K que se emplea con el siguiente metodo:

```python
#Model Complexity Code with K range (1,50)
#En los modelos previamente realizados se empleo n_neighbors(K) con rangos de 1 a 50

neig= np.arange(1,50)
train_accurary=[]
test_accuracy=[]

#Bucle para diferentes valores de K

for i, K in enumerate(neig):
    
    #K from 1 a 50
    knn= KNeighborsClassifier (n_neighbors=K)
    
    #Ajustar con Knn (metodo fit)
    knn.fit(X_train, Y_train)
    
    #Train Accuracy
    train_accurary.append(knn.score(X_train,Y_train))
    
    #test_accuracy
    test_accuracy.append(knn.score(X_test,Y_test))
    
print (f"Datos Entrenamiendo de Exactitud:\n{(train_accurary)}")

print (f"Datos Testeo de Exactitud:\n{(test_accuracy)}")
```

```python
#Plot Code

plt.figure(figsize=[13,8])

plt.plot(neig, test_accuracy, label="Testing Accuracy")
plt.plot(neig, train_accurary, label="Training Accuracy")

plt.legend()
plt.title("Número de Vecinos (K) con respecto a Precisión de Pruebas y Precisión de Entrenamiento")
plt.xlabel("Number of Neightbors")
plt.ylabel("Accuracy")
plt.xticks(neig)

plt.show()

print ("Best Accuracy is {} with K={}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
